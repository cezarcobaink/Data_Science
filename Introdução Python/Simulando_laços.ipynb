```python
class Aluno:
    def __init__(self, nome, cpf, email, matricula):
        # Inicializa os atributos do aluno com os dados fornecidos
        self.nome = nome
        self.cpf = cpf
        self.email = email
        self.matricula = matricula
        self.notas = []

    def adicionar_nota(self, nota):
        # Adiciona uma nota à lista de notas do aluno
        self.notas.append(nota)

    def calcular_media(self):
        # Calcula a média das notas do aluno
        if len(self.notas) == 0:
            return 0
        return sum(self.notas) / len(self.notas)


def imprimir_dados_aluno(aluno):
    # Imprime todos os dados do aluno, incluindo as notas e a média
    print("\nDados do aluno:")
    print(f"Nome: {aluno.nome}")
    print(f"CPF: {aluno.cpf}")
    print(f"Email: {aluno.email}")
    print(f"Matrícula: {aluno.matricula}")
    print(f"Notas: {aluno.notas}")
    print(f"Média: {aluno.calcular_media()}")


def main():
    alunos = []  # Lista para armazenar os alunos cadastrados

    while True:
        # Menu de opções
        print("\n1. Cadastrar aluno")
        print("2. Lançar notas")
        print("3. Imprimir dados de aluno")
        print("4. Sair")
        opcao = int(input("Escolha uma opção: "))

        if opcao == 1:
            # Cadastrar um novo aluno
            nome = input("Nome do aluno: ")
            cpf = input("CPF do aluno: ")
            email = input("Email do aluno: ")
            matricula = input("Matrícula do aluno: ")
            alunos.append(Aluno(nome, cpf, email, matricula))
        elif opcao == 2:
            # Lançar notas para um aluno
            matricula = input("Matrícula do aluno: ")
            # Encontrar o aluno pela matrícula
            aluno_encontrado = next((aluno for aluno in alunos if aluno.matricula == matricula), None)

            if aluno_encontrado:
                # Adicionar três notas ao aluno
                for _ in range(3):
                    nota = float(input(f"Nota {_ + 1} do aluno: "))
                    aluno_encontrado.adicionar_nota(nota)
                # Calcular a média das notas
                media = aluno_encontrado.calcular_media()
                if media >= 6:
                    print(f"Aluno {aluno_encontrado.nome}, você foi aprovado!")
                    imprimir_dados_aluno(aluno_encontrado)
                else:
                    print("Média insuficiente. Lançando nota adicional.")
                    # Adicionar uma nota adicional se a média for insuficiente
                    nota_adicional = float(input("Nota adicional: "))
                    aluno_encontrado.adicionar_nota(nota_adicional)
                    nova_media = aluno_encontrado.calcular_media()
                    if nova_media >= 6:
                        print(f"Aluno {aluno_encontrado.nome}, você foi aprovado!")
                        imprimir_dados_aluno(aluno_encontrado)
                    else:
                        print(f"Aluno {aluno_encontrado.nome}, você foi reprovado.")
            else:
                print("Aluno não encontrado.")
        elif opcao == 3:
            # Imprimir os dados de um aluno
            matricula = input("Matrícula do aluno: ")
            # Encontrar o aluno pela matrícula
            aluno_encontrado = next((aluno for aluno in alunos if aluno.matricula == matricula), None)

            if aluno_encontrado:
                imprimir_dados_aluno(aluno_encontrado)
            else:
                print("Aluno não encontrado.")
        elif opcao == 4:
            # Sair do programa
            break
        else:
            print("Opção inválida.")


if __name__ == "__main__":
    main()
```
